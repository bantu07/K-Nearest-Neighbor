```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

```python
df = pd.read_csv("C://Users//Dell//Downloads//Excel R//R codes//knn//wbcd.csv")
df = df.drop('id',axis=1)
df.head()
```

```python
dum = pd.get_dummies(df.diagnosis,columns=({'B':1,'M':0}),drop_first=True)
dum
```

```python
df_d = pd.concat(objs=(df,dum),axis=1)
df_d = df_d.drop('diagnosis',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df_d.drop('M',axis= 1))
scaled_df = scaler.transform(df_d.drop('M',axis = 1))
```

```python
scaled_df = pd.DataFrame(data=scaled_df,columns=df.columns[1:])
scaled_df
```

```python
X = scaled_df.iloc[:,:31]
y = df_d['M']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
classifier = knn.fit(X_train,y_train)
pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
